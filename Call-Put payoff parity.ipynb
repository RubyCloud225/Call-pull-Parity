```python
import matplotlib.pyplot as plt
%pylab inline
price = np.arange(900,1000,1)
strike = 940
premium = 7.5
payoff = [max(-premium,i - strike-premium) for i in price]
```

```python
plt.plot(price, payoff)
plt.xlabel('price at T S_T ($)')
plt.ylabel('payoff')
plt.title('Call option Payoff at Expiry')
plt.grid(True)
```
